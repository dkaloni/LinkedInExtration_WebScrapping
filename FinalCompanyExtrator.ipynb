### Profile Link Extraction

```python
def getURL():
    page_sources = BeautifulSoup(browser.page_source, 'lxml')
    profiles = page_sources.find_all('a',class_ ="app-aware-link")
    all_profile_URL = []
    for profile in profiles:
        profile_ID = profile.get('href')
        profile_URL = profile_ID
        if profile_URL not in all_profile_URL and "SHARED" not in profile_URL and "ACoAA" not in profile_URL and "keyword" not in profile_URL:
            all_profile_URL.append(profile_URL)
            print(profile_URL)
    return all_profile_URL
print(getURL())
```

```python
URLs_all_page = []
```

```python
def getURLsonPages():
    number_of_page =50
    URLs_all_page =[]
    for page in range(number_of_page):
        URLs_one_page = getURL()
        browser.execute_script('window.scrollTo(0,document.body.scrollHeight);')
        sleep(2)
        next_button = browser.find_element_by_class_name("artdeco-pagination__button--next")
        next_button.click()
        URLs_all_page += URLs_one_page
        sleep(2)
    return URLs_all_page
listforpages = getURLsonPages()
```
